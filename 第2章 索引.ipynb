## 一、单级索引
### 1. loc方法、iloc方法、[]操作符
最常用的索引方法可能就是这三类，其中iloc表示位置索引，loc表示标签索引，[]也具有很大的便利性，各有特点

#### ① 单行索引

```python
df.loc[1103]
```

#### ② 多行索引

```python
df.loc[[1102,2304]]
```

注意：所有在loc中使用的切片**全部包含右端点**！这是因为如果作为Pandas的使用者，那么肯定不太关心最后一个标签再往后一位是什么，但是如果是左闭右开，那么就很麻烦，先要知道再后面一列的名字是什么，非常不方便，因此Pandas中将loc设计为左右全闭。

```python
df.loc[1304:2103].head()
```

```python
df.loc[2402::-1].head()
```

### 2. 布尔索引
#### （a）布尔符号
'&','|','~'：分别代表和and，或or，取反not

```python
df[(df['Gender']=='F')&(df['Address']=='street_2')].head()
```

```python
df[(df['Math']>85)|(df['Address']=='street_7')].head()
```

```python
df[~((df['Math']>75)|(df['Address']=='street_1'))].head()
```

loc和[]中相应位置都能直接使用布尔Series选择，而iloc方法需要Series.values()，即转为布尔列表：

```python
df.loc[df['Math']>60,df.columns=='Physics'].head()
```

思考：为什么`df.loc[df['Math']>60,(df[:8]['Address']=='street_6').values].head()`得到和上述结果一样？values能去掉吗？

A：如下所示，`df.columns=='Physics'`得到了选取列的布尔列表，除了最后一个均为False，而`(df[:8]['Address']=='street_6').values`虽然和列无关系，但是得到的布尔列表和上一个相同，因此得到的结果一样。values不能去掉，因为这样是针对索引进行的筛选，但是在代码中应该是对列名进行筛选，去掉会报错。

```python
df.columns=='Physics'
```

```python
(df[:8]['Address']=='street_6').values
```

### 4. 区间索引
此处介绍并不是说只能在单级索引中使用区间索引，只是作为一种特殊类型的索引方式，在此处先行介绍
#### （a）利用interval_range方法

```python
pd.interval_range(start=0,end=5)
#closed参数可选'left''right''both''neither'，默认左开右闭
```

```python
pd.interval_range(start=0,periods=8,freq=5)
#periods参数控制区间个数，freq控制步长
```

#### （b）利用cut将数值列转为区间为元素的分类变量
例如统计数学成绩的区间情况：
> 注意，如果没有类型转换，此时并不是区间类型，而是category类型

```python
math_interval = pd.cut(df['Math'],bins=[0,40,60,80,100])
math_interval.head()
```

#### （c）区间索引的选取

```python
df_i = df.join(math_interval,rsuffix='_interval')[['Math','Math_interval']]\
            .reset_index().set_index('Math_interval')
# reset_index()表示重置index并保留原有index，若设置drop=True，则删掉原有index
df_i.head()
```

```python
df_i.loc[65]
#包含该值的区间就会被选中
```

```python
df_i.loc[[65,90]]
```

如果想要选取某个区间，先要把分类变量转为区间变量，再使用overlaps方法：

```python
#df_i.loc[pd.Interval(70,75)].head() 报错
df_i[df_i.index.astype('interval').overlaps(pd.Interval(70, 85))]
```

```python
df_using_mul.head()
```

#### （a）一般切片
当索引不排序时，单个索引会报出性能警告，可以使用函数`df.index.is_lexsorted()`检查索引是否排序

### 3. 多层索引中的slice对象

```python
L1,L2 = ['A','B','C'],['a','b','c']
mul_index1 = pd.MultiIndex.from_product([L1,L2],names=('Upper', 'Lower'))
L3,L4 = ['D','E','F'],['d','e','f']
mul_index2 = pd.MultiIndex.from_product([L3,L4],names=('Big', 'Small'))
df_s = pd.DataFrame(np.random.rand(9,9),index=mul_index1,columns=mul_index2)
df_s
```

```python
idx=pd.IndexSlice # 创建一个便于多层索引切片的对象
```

索引Slice的使用非常灵活：

```python
df_s['D']['d']>0.3
```

```python
df_s.sum()>4
```

```python
df_s.loc[idx['B':,df_s['D']['d']>0.3],idx[df_s.sum()>4]]
# 对于loc中的行筛选，因为有多层索引，则idx[,]中，第一块为第一层筛选，第二块为利用列筛选进行第二层筛选
# 对于loc中的列筛选，第二层的df_s.sum()默认为对列迭代对行求和求和，因此返回一个长度为9的数值列表，第一层则可以省略
```

```python
# 多行查询
%timeit df.loc[lambda x:x['Address']=="street_4"]
%timeit df.loc[df['Address'].isin(["street_4"])]
%timeit df.loc[[True if i == "street_4" else False for i in df['Address']]]  # 测试下来这个最快
%timeit df.iloc[(df['Address']=="street_4").values]
%timeit df[df['Address'] == "street_4"]
%timeit df.query('Address == "street_4"')
```

```python
# 多列查询
%timeit df.loc[:,['Address','School']]
%timeit df[['Address','School']]
```

#### （d）请问六项种族指标（HP、物攻、特攻、物防、特防、速度）极差的均值最大的是哪个属性（只考虑第一属性，且均值是对属性而言）？
注意审题，指的是各个指标之间的极差。

```python
# 使用分组方法
columns = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']
df2['range'] = df2[columns].max(axis=1) - df2[columns].min(axis=1)
df2.groupby('Type 1')['range'].mean().idxmax()
```

```python
# 不使用分组方法
Types = df2['Type 1'].unique()
columns = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']
df2['range'] = df2[columns].max(axis=1) - df2[columns].min(axis=1)
maxnum = 0
result = ''
for i in Types:
    temp = df2.loc[df2['Type 1']==i,'range'].mean()
    if temp > maxnum:
        maxnum = temp
        result = i
result
```
